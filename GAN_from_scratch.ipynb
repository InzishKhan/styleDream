```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, random_split
from torchvision import transforms
import torchvision.utils as vutils
from datasets import load_dataset
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Generator with Upsample + Conv2D
class Generator(nn.Module):
    def __init__(self, latent_dim):
        super(Generator, self).__init__()
        self.init_size = 8
        self.l1 = nn.Sequential(nn.Linear(latent_dim, 256 * self.init_size ** 2))
        self.conv_blocks = nn.Sequential(
            nn.BatchNorm2d(256),
            nn.Upsample(scale_factor=2),  # 8x8 -> 16x16
            nn.Conv2d(256, 128, 3, stride=1, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(True),
            nn.Upsample(scale_factor=2),  # 16x16 -> 32x32
            nn.Conv2d(128, 64, 3, stride=1, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(True),
            nn.Upsample(scale_factor=2),  # 32x32 -> 64x64
            nn.Conv2d(64, 3, 3, stride=1, padding=1),
            nn.Tanh()
        )

    def forward(self, z):
        out = self.l1(z).view(z.size(0), 256, self.init_size, self.init_size)
        return self.conv_blocks(out)

# Discriminator with Spectral Normalization
class Discriminator(nn.Module):
    def __init__(self):
        super(Discriminator, self).__init__()
        def block(in_channels, out_channels):
            return nn.Sequential(
                nn.utils.spectral_norm(nn.Conv2d(in_channels, out_channels, 4, stride=2, padding=1)),
                nn.LeakyReLU(0.2, inplace=True),
                nn.Dropout(0.25)
            )
        self.model = nn.Sequential(
            block(3, 64),
            block(64, 128),
            block(128, 256),
            block(256, 512),
            nn.Flatten(),
            nn.Linear(512 * 4 * 4, 1),
            nn.Sigmoid()
        )

    def forward(self, img):
        return self.model(img)

# Weight initialization
def weights_init_normal(m):
    classname = m.__class__.__name__
    if classname.find("Conv") != -1 or classname.find("Linear") != -1:
        nn.init.xavier_normal_(m.weight)
        if m.bias is not None:
            nn.init.constant_(m.bias, 0)

# Training
def train_gan(generator, discriminator, dataloader, latent_dim, num_epochs=100, lr=0.0002):
    criterion = nn.BCELoss()
    optimizer_g = optim.Adam(generator.parameters(), lr=lr, betas=(0.5, 0.999))
    optimizer_d = optim.Adam(discriminator.parameters(), lr=lr, betas=(0.5, 0.999))

    generator.apply(weights_init_normal)
    discriminator.apply(weights_init_normal)
    generator.to(device)
    discriminator.to(device)

    for epoch in range(num_epochs):
        for imgs in dataloader:
            imgs = imgs.to(device)
            batch_size = imgs.size(0)

            valid = torch.ones(batch_size, 1, device=device)
            fake = torch.zeros(batch_size, 1, device=device)

            # Training Generator
            optimizer_g.zero_grad()
            z = torch.randn(batch_size, latent_dim, device=device)
            gen_imgs = generator(z)
            g_loss = criterion(discriminator(gen_imgs), valid)
            g_loss.backward()
            optimizer_g.step()

            # Training Discriminator
            optimizer_d.zero_grad()
            real_loss = criterion(discriminator(imgs), valid)
            fake_loss = criterion(discriminator(gen_imgs.detach()), fake)
            d_loss = (real_loss + fake_loss) / 2
            d_loss.backward()
            optimizer_d.step()

        if epoch % 10 == 0:
            print(f"[Epoch {epoch}/{num_epochs}] Loss D: {d_loss:.4f}, Loss G: {g_loss:.4f}")

    return generator

# Loading the anime-style dataset
dataset = load_dataset("Lifeinhockey/man_animestyle")
transform = transforms.Compose([
    transforms.Resize((64, 64)),
    transforms.ToTensor(),
    transforms.Normalize([0.5]*3, [0.5]*3)
])
images = []
for example in dataset['train']:
    img = example['image']
    if isinstance(img, np.ndarray):
        img = Image.fromarray(img)
    img = img.convert("RGB")
    img = transform(img)
    images.append(img)
full_dataset = torch.stack(images)

# Splitting
train_size = int(0.9 * len(full_dataset))
train_dataset, _ = random_split(full_dataset, [train_size, len(full_dataset) - train_size])
train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)

# models
latent_dim = 100
generator = Generator(latent_dim)
discriminator = Discriminator()
trained_generator = train_gan(generator, discriminator, train_loader, latent_dim, num_epochs=10000)

# Generate samples
z = torch.randn(25, latent_dim, device=device)
fake_imgs = trained_generator(z).detach().cpu()
fake_imgs = (fake_imgs + 1) / 2  # Rescale to [0, 1]

# Showing grid of images
grid = vutils.make_grid(fake_imgs, nrow=5)
plt.figure(figsize=(10, 10))
plt.imshow(grid.permute(1, 2, 0))
plt.axis("off")
plt.title("Generated Anime-Style Faces")
plt.show()
```
